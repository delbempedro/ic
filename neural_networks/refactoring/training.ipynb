```python
number_of_runs = 1
number_of_shots = 1024
grid_grain = 13

for number_of_bits in range(2, 3):
    max_iterations = grid_grain ** number_of_bits

    with open("results.txt", "w") as f:
        f.write(f"\nnumber_of_bits: {number_of_bits}\n")
        f.write("  exaustive search:\n")

        # Exaustiva - single qubit
        f.write("    All inputs in same qubit:\n")
        angles, error, iterations, history_list = trainer_qNN.single_qubit_qNN_exaustive_search(
            inputs, expected_outputs, number_of_bits=number_of_bits,
            number_of_runs=number_of_runs, number_of_shots=number_of_shots,
            number_of_inputs_per_qubit=2, save_history=True)
        f.write(f"      Final Iteration: {iterations}, Error: {error}\n")
        for idx, err in enumerate(history_list, 1):
            f.write(f"        Iteration: {idx}, Error: {err}\n")

        # Exaustiva - multi qubit
        f.write("    One input per qubit:\n")
        angles, error, iterations, history_list = trainer_qNN.multi_qubit_qNN_exaustive_search(
            inputs, expected_outputs, number_of_bits=number_of_bits,
            number_of_runs=number_of_runs, number_of_shots=number_of_shots,
            save_history=True)
        f.write(f"      Final Iteration: {iterations}, Error: {error}\n")
        for idx, err in enumerate(history_list, 1):
            f.write(f"        Iteration: {idx}, Error: {err}\n")

        # Gradient Descent
        f.write("  gradient descent:\n")

        f.write("    All inputs in same qubit:\n")
        angles, error, iterations, history_list = trainer_qNN.single_qubit_qNN_gradient_descent(
            inputs, expected_outputs, number_of_bits=number_of_bits,
            number_of_runs=number_of_runs, number_of_shots=number_of_shots,
            max_iterations=max_iterations, learning_rate=0.1,
            number_of_inputs_per_qubit=2, save_history=True)
        f.write(f"      Final Iteration: {iterations}, Error: {error}\n")
        for idx, err in enumerate(history_list, 1):
            f.write(f"        Iteration: {idx}, Error: {err}\n")

        f.write("    One input per qubit:\n")
        angles, error, iterations, history_list = trainer_qNN.multi_qubit_qNN_gradient_descent(
            inputs, expected_outputs, number_of_bits=number_of_bits,
            number_of_runs=number_of_runs, number_of_shots=number_of_shots,
            max_iterations=max_iterations, learning_rate=0.1,
            save_history=True)
        f.write(f"      Final Iteration: {iterations}, Error: {error}\n")
        for idx, err in enumerate(history_list, 1):
            f.write(f"        Iteration: {idx}, Error: {err}\n")

        # Simulated Annealing
        f.write("  simulated annealing:\n")

        f.write("    All inputs in same qubit:\n")
        angles, error, iterations, history_list = trainer_qNN.single_qubit_qNN_simulated_annealing(
            inputs, expected_outputs, number_of_bits=number_of_bits,
            number_of_runs=number_of_runs, number_of_shots=number_of_shots,
            max_iterations=max_iterations, number_of_inputs_per_qubit=2,
            save_history=True)
        f.write(f"      Final Iteration: {iterations}, Error: {error}\n")
        for idx, err in enumerate(history_list, 1):
            f.write(f"        Iteration: {idx}, Error: {err}\n")

        f.write("    One input per qubit:\n")
        angles, error, iterations, history_list = trainer_qNN.multi_qubit_qNN_simulated_annealing(
            inputs, expected_outputs, number_of_bits=number_of_bits,
            number_of_runs=number_of_runs, number_of_shots=number_of_shots,
            max_iterations=max_iterations, save_history=True)
        f.write(f"      Final Iteration: {iterations}, Error: {error}\n")
        for idx, err in enumerate(history_list, 1):
            f.write(f"        Iteration: {idx}, Error: {err}\n")

        # Random Search
        f.write("  random search:\n")

        f.write("    All inputs in same qubit:\n")
        angles, error, iterations, history_list = trainer_qNN.single_qubit_qNN_random_search(
            inputs, expected_outputs, number_of_bits=number_of_bits,
            number_of_runs=number_of_runs, number_of_shots=number_of_shots,
            max_iterations=max_iterations, number_of_inputs_per_qubit=2,
            save_history=True)
        f.write(f"      Final Iteration: {iterations}, Error: {error}\n")
        for idx, err in enumerate(history_list, 1):
            f.write(f"        Iteration: {idx}, Error: {err}\n")

        f.write("    One input per qubit:\n")
        angles, error, iterations, history_list = trainer_qNN.multi_qubit_qNN_random_search(
            inputs, expected_outputs, number_of_bits=number_of_bits,
            number_of_runs=number_of_runs, number_of_shots=number_of_shots,
            max_iterations=max_iterations, save_history=True)
        f.write(f"      Final Iteration: {iterations}, Error: {error}\n")
        for idx, err in enumerate(history_list, 1):
            f.write(f"        Iteration: {idx}, Error: {err}\n")

        # Genetic Algorithm
        f.write("  genetic algorithm:\n")
        f.write("    All inputs in same qubit:\n")
        angles, error, iterations, history_list = trainer_qNN.single_qubit_qNN_genetic_algorithm(
            inputs, expected_outputs, number_of_bits=number_of_bits,
            number_of_runs=number_of_runs, number_of_shots=number_of_shots, number_of_inputs_per_qubit=2,
            save_history=True)
        f.write(f"      Final Iteration: {iterations}, Error: {error}\n")
        for idx, err in enumerate(history_list, 1):
            f.write(f"        Iteration: {idx}, Error: {err}\n")
        f.write("    One input per qubit:\n")
        angles, error, iterations, history_list = trainer_qNN.multi_qubit_qNN_genetic_algorithm(
            inputs, expected_outputs, number_of_bits=number_of_bits,
            number_of_runs=number_of_runs, number_of_shots=number_of_shots, save_history=True)
        f.write(f"      Final Iteration: {iterations}, Error: {error}\n")
        for idx, err in enumerate(history_list, 1):
            f.write(f"        Iteration: {idx}, Error: {err}\n")
```
